```python
# ── CELL 1: Install all dependencies ─────────────────────────
import sys
!{sys.executable} -m pip install requests beautifulsoup4 lxml pandas sqlite3 schedule

# ── CELL 2: Imports ──────────────────────────────────────────
import requests
from bs4 import BeautifulSoup
import sqlite3
import pandas as pd
import smtplib
import json
import datetime
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart
import warnings
warnings.filterwarnings("ignore")

print("All imports successful!")

# ── CELL 3: Define competitor list ───────────────────────────
COMPETITORS = [
    {
        "name": "Physics Wallah",
        "website": "https://www.pw.live",
        "careers_url": "https://www.pw.live/careers",
        "news_query": "Physics Wallah 2026"
    },
    {
        "name": "Unacademy",
        "website": "https://unacademy.com",
        "careers_url": "https://unacademy.com/careers",
        "news_query": "Unacademy 2026"
    },
    {
        "name": "Vedantu",
        "website": "https://www.vedantu.com",
        "careers_url": "https://www.vedantu.com/careers",
        "news_query": "Vedantu 2026"
    },
    {
        "name": "upGrad",
        "website": "https://www.upgrad.com",
        "careers_url": "https://www.upgrad.com/careers",
        "news_query": "upGrad 2026"
    },
    {
        "name": "Scaler",
        "website": "https://www.scaler.com",
        "careers_url": "https://www.scaler.com/careers",
        "news_query": "Scaler Academy 2026"
    },
]

print(f"Tracking {len(COMPETITORS)} competitors:")
for c in COMPETITORS:
    print(f"  - {c['name']}")

# ── CELL 4: Web scraper function ─────────────────────────────
HEADERS = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                  "AppleWebKit/537.36 (KHTML, like Gecko) "
                  "Chrome/120.0.0.0 Safari/537.36"
}

def scrape_website(url):
    """Scrape basic signals from a competitor website"""
    try:
        response = requests.get(url, headers=HEADERS, timeout=10)
        soup = BeautifulSoup(response.text, "lxml")

        # Extract page title
        title = soup.title.string.strip() if soup.title else "N/A"

        # Count internal links (proxy for content depth)
        links = soup.find_all("a", href=True)
        internal_links = [l for l in links if url.split("/")[2] in str(l.get("href", ""))]

        # Extract meta description
        meta = soup.find("meta", attrs={"name": "description"})
        description = meta["content"].strip() if meta and meta.get("content") else "N/A"

        # Count images (proxy for visual content richness)
        images = len(soup.find_all("img"))

        return {
            "title": title[:100],
            "description": description[:200],
            "internal_links": len(internal_links),
            "image_count": images,
            "status": "success"
        }
    except Exception as e:
        return {
            "title": "N/A",
            "description": "N/A",
            "internal_links": 0,
            "image_count": 0,
            "status": f"error: {str(e)[:50]}"
        }

def scrape_careers(url):
    """Count job postings as a hiring activity signal"""
    try:
        response = requests.get(url, headers=HEADERS, timeout=10)
        soup = BeautifulSoup(response.text, "lxml")

        # Common job listing keywords
        job_keywords = ["engineer", "manager", "analyst", "designer",
                        "developer", "marketing", "sales", "intern"]
        text = soup.get_text().lower()
        job_count = sum(text.count(kw) for kw in job_keywords)

        return min(job_count, 999)  # cap at 999
    except:
        return 0

# ── CELL 5: Test scraper on one competitor ───────────────────
print("Testing scraper on Physics Wallah...")
result = scrape_website(COMPETITORS[0]["website"])
for k, v in result.items():
    print(f"  {k}: {v}")

# ── CELL 6: Setup SQLite database ────────────────────────────
conn = sqlite3.connect("competitor_tracker.db")
cursor = conn.cursor()

# Create main tracking table
cursor.execute("""
CREATE TABLE IF NOT EXISTS competitor_data (
    id              INTEGER PRIMARY KEY AUTOINCREMENT,
    scraped_date    TEXT NOT NULL,
    company_name    TEXT NOT NULL,
    website         TEXT,
    page_title      TEXT,
    meta_description TEXT,
    internal_links  INTEGER,
    image_count     INTEGER,
    job_signal      INTEGER,
    scrape_status   TEXT
)
""")

# Create news table
cursor.execute("""
CREATE TABLE IF NOT EXISTS competitor_news (
    id              INTEGER PRIMARY KEY AUTOINCREMENT,
    scraped_date    TEXT NOT NULL,
    company_name    TEXT NOT NULL,
    headline        TEXT,
    source          TEXT
)
""")

conn.commit()
print("Database created successfully!")
print("Tables: competitor_data, competitor_news")

# ── CELL 7: Run full scrape & store in DB ────────────────────
today = datetime.date.today().isoformat()

print(f"\nScraping all competitors for {today}...")
print("-" * 50)

for company in COMPETITORS:
    print(f"Scraping {company['name']}...", end=" ")

    # Scrape website
    web_data  = scrape_website(company["website"])
    job_signal = scrape_careers(company["careers_url"])

    # Insert into database
    cursor.execute("""
        INSERT INTO competitor_data
        (scraped_date, company_name, website, page_title,
         meta_description, internal_links, image_count,
         job_signal, scrape_status)
        VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?)
    """, (
        today,
        company["name"],
        company["website"],
        web_data["title"],
        web_data["description"],
        web_data["internal_links"],
        web_data["image_count"],
        job_signal,
        web_data["status"]
    ))

    conn.commit()
    print(f"Done! (Status: {web_data['status']})")

print("-" * 50)
print("All competitors scraped and stored!")

# ── CELL 8: SQL queries to analyse the data ──────────────────
print("\n=== COMPETITOR INTELLIGENCE REPORT ===\n")

# Query 1: Overview of all companies today
print("1. Today's Snapshot:")
df1 = pd.read_sql_query("""
    SELECT
        company_name        AS Company,
        internal_links      AS Internal_Links,
        image_count         AS Images,
        job_signal          AS Hiring_Activity,
        scrape_status       AS Status
    FROM competitor_data
    WHERE scraped_date = ?
    ORDER BY job_signal DESC
""", conn, params=(today,))
print(df1.to_string(index=False))

# Query 2: Who is hiring the most
print("\n2. Hiring Activity Ranking (higher = more active hiring):")
df2 = pd.read_sql_query("""
    SELECT
        company_name    AS Company,
        job_signal      AS Hiring_Score
    FROM competitor_data
    WHERE scraped_date = ?
    ORDER BY job_signal DESC
""", conn, params=(today,))
print(df2.to_string(index=False))

# Query 3: Content depth ranking
print("\n3. Content Depth Ranking (internal links as proxy):")
df3 = pd.read_sql_query("""
    SELECT
        company_name        AS Company,
        internal_links      AS Links,
        image_count         AS Images,
        (internal_links + image_count) AS Content_Score
    FROM competitor_data
    WHERE scraped_date = ?
    ORDER BY Content_Score DESC
""", conn, params=(today,))
print(df3.to_string(index=False))

# ── CELL 9: Generate weekly digest email ─────────────────────
def generate_digest(conn, today):
    """Generate a plain text digest from the database"""

    df = pd.read_sql_query("""
        SELECT company_name, internal_links, image_count,
               job_signal, scrape_status
        FROM competitor_data
        WHERE scraped_date = ?
        ORDER BY job_signal DESC
    """, conn, params=(today,))

    lines = []
    lines.append("=" * 50)
    lines.append("VIORA EDTECH COMPETITOR DIGEST")
    lines.append(f"Week of {today}")
    lines.append("=" * 50)
    lines.append("")
    lines.append("HIRING ACTIVITY (Top Signal This Week):")
    lines.append("-" * 40)

    for _, row in df.iterrows():
        lines.append(
            f"{row['company_name']:<20} Hiring Score: {row['job_signal']}"
        )

    lines.append("")
    lines.append("CONTENT DEPTH RANKING:")
    lines.append("-" * 40)

    df["content_score"] = df["internal_links"] + df["image_count"]
    df_sorted = df.sort_values("content_score", ascending=False)

    for _, row in df_sorted.iterrows():
        lines.append(
            f"{row['company_name']:<20} Score: {row['content_score']}"
        )

    lines.append("")
    lines.append("KEY INSIGHT:")
    top_hiring = df.iloc[0]["company_name"]
    lines.append(
        f"{top_hiring} shows highest hiring activity this week "
        f"suggesting aggressive expansion. Monitor closely."
    )

    lines.append("")
    lines.append("=" * 50)
    lines.append("Powered by Competitor Intelligence Tracker")
    lines.append("github.com/aastha688")

    return "\n".join(lines)

digest = generate_digest(conn, today)
print(digest)

# Save digest to file
with open("weekly_digest.txt", "w") as f:
    f.write(digest)
print("\nweekly_digest.txt saved!")

# ── CELL 10: Send digest via email ───────────────────────────
def send_digest_email(digest_text, sender_email, sender_password, receiver_email):
    """Send digest via Gmail"""
    try:
        msg = MIMEMultipart()
        msg["From"]    = sender_email
        msg["To"]      = receiver_email
        msg["Subject"] = f"EdTech Competitor Digest — {datetime.date.today()}"

        msg.attach(MIMEText(digest_text, "plain"))

        server = smtplib.SMTP("smtp.gmail.com", 587)
        server.starttls()
        server.login(sender_email, sender_password)
        server.sendmail(sender_email, receiver_email, msg.as_string())
        server.quit()

        print("Email sent successfully!")
    except Exception as e:
        print(f"Email error: {e}")
        print("Tip: Use Gmail App Password, not your main password")

# ── Fill in your details and uncomment to send ───────────────
# send_digest_email(
#     digest,
#     sender_email="your_gmail@gmail.com",
#     sender_password="your_app_password",   # Gmail App Password
#     receiver_email="mishraaastha017@gmail.com"
# )

print("\nAll done!")
print("Files generated:")
print("  - competitor_tracker.db (SQLite database)")
print("  - weekly_digest.txt (this week's report)")
conn.close()
```
